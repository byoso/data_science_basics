## Set up

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# let's build some datas for the example

m1 = np.array([
    ["no","factor1",0],
    ["yes","factor2",2],
    ["no","factor3",0], 
    ["no","factor4",2],
    ["yes","factor1", 1], 
    ["no","factor2",3],
    ["yes","factor5",3],
    ["no","factor1",0],
    ["yes","factor5",3],
    ["yes","factor1",2],
    ["yes","factor5",1],
    ["no","factor1",2],
    ["no","factor2",0],
    ["yes","factor1",1],
    ["no","factor3",2]
])

# create a dataframe with columns
df1 = pd.DataFrame(m1, columns=["Request","Factor", "Impact"])
df1["Impact"] = df1["Impact"].astype(int)

a1 = np.random.uniform(
    low=0,
    high=10,
    size=(m1.shape[0], 1)
)
df1["new_column"] = a1


df1

````

## First pyplot

### Plot

````python
# here we create a new column for a better example
df1["x"] = df1["Impact"].apply(lambda x: x * np.random.randint(1, 5))

## First sort the values for the x axis
df1.sort_values(
    by=["x"], # the criteria
    ascending=True, # the order
    inplace=True # True: modify the df, False: doesn't modify the original df, used if you store the result in a new df.
) 
df1
````

````python
# accroding to the official doc:

plt.style.use('_mpl-gallery')

# make data
x = df1["x"]
y = df1["new_column"]

# plot
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(
    x, y,
    linewidth=2.0, color="red", marker="x" # these are optionnal
)
ax.set_xlabel('x Impact')  # Add an x-label to the axes.
ax.set_ylabel('new_column')  # Add a y-label to the axes.
ax.set_title("New column in function of 'x impact' ") 


plt.show()
````
